```python
# State | on 0 | on 1 
# Sources: http://turbotm.de/~heiner/BB/bbsimtab.html
rules=open("Transitions-tables/test.txt",'r').read()
moves={'L':-1,'R':1}
```

```python
print(rules)
rules=[x.split() for x in rules.splitlines()]
```

```python
def rules_output(l,on,rules_table):
    for x in rules_table:
        if l in x[0]:
            return tuple(x[int(on)+1])           
```

```python
# Verifying output
rules_output('A','0',rules)
```

```python
def create_env(x):
    return ['.']*x,x//2
ret,index=create_env(10)
```

```python
# 'L'/'R','0'/'1'

def write(move,pri):
    global index
    ret[index]=pri
    index+=moves[move]
    # Auto-extend
    if index<0 or index>len(ret)-1:
        ret.insert(0,'.')
        ret.append('.')
        index+=1
    if ret[index]=='.':
        ret[index]='0'
#  instructions       array
#
#                         0
# write('L','1')         01
# write('L','1')        011
# write('L','1')       0111
# write('R','1')       1111 
# write('R','1')       1111
# write('R','1')       1111
# write('R','1')       11110
# write('R','1')       111110
# write('L','0')       111100
```

```python
def init_stats(table_rules):
    stats={}
    for x in table_rules:
        stats[x[0]]=0
        for i in range(len(x)-1):
            stats[x[0]+str(i)]=0
    return stats
```

```python
def main(steps):
    stats=init_stats(rules)
    # Initialize first values
    ret[index],State='0','A'
    print("Step State Output\n")
    for x in range(steps+1):
        print("{}     {} {}".format((4-len(str(x)))*' '+str(x),State,"".join(ret).replace('.','')))
        # Calculating states amount
        stats[State+ret[index]]+=1
        stats[State]+=1
        # Output from transition table
        rr=rules_output(State,ret[index],rules)
        write(rr[2],rr[1])
        State=rr[0]
    return stats,"".join(ret).replace('.',''),State

stats,out,state=main(200)
```

```python
def parse_stats(d):
    print("State Count\n")
    for k,v in d.items():
        print("{}    {}".format((5-len(k))*' '+k,v))
parse_stats(stats)
```
